```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class Trajectory():
    def __init__(self):
        name           = ""
        Mean_diff      = 0 
        Chi_Step       = 0
        Chi_Line       = 0
        Step_Time      = 0
        Times          = np.array([])
        Signal         = np.array([])
        Signal_er      = np.array([])
        Background     = np.array([])
        Background_er  = np.array([])
```

```python
Traj = np.load('Trajectory.npy').item()
```

```python
KEYS = Traj.keys()
len(KEYS)
```

```python
MD  = []
CHI = []
for key in KEYS:
    MD.append(Traj[key].Mean_diff)
    cc = Traj[key].Chi_Step - Traj[key].Chi_Line
    CHI.append(cc)
```

```python
plt.scatter(CHI,MD)
```

```python
qq=0
for key in KEYS:
    md = Traj[key].Mean_diff
    cc = Traj[key].Chi_Step - Traj[key].Chi_Line
    if cc < -1 and md>5:  
        times    = np.arange(0,len(Traj[key].Signal))*2.5
        qq+=1
        print(key)
        print(md)
        print(cc)
        #print(spots[x])
        plt.scatter(times,Traj[key].Signal)
        plt.scatter(times,Traj[key].Background)
        plt.errorbar(times,Traj[key].Signal    ,yerr=Traj[key].Signal_er,color='r',fmt='.')
        plt.errorbar(times,Traj[key].Background,yerr=Traj[key].Background_er,color='k',fmt='.')
        plt.axvline(Traj[key].Step_Time*2.5)
        plt.show()   
```
